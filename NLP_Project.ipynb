"Performing Sentiment Analysis on IMDB Movie Reviews Kaggle Dataset Using NLP (Natural Language Processing)"

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier #the random forest classifier can be used to solve for regression or classification problems.
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import roc_auc_score
from bs4 import BeautifulSoup
import re
import nltk   #NLTK is a toolkit build for working with NLP in Python
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords
```

```python
df=pd.read_csv('/content/drive/MyDrive/NLPlabeledData.tsv',delimiter="\t", quoting=3) #çift tırnakdan kurtulmak için
```

```python
df.head()
```

```python
len(df)
#len(df['review'])
```

```python
nltk.download('stopwords')
```

```python
def process(review):
  # review without HTML tags
  review=BeautifulSoup(review).get_text()
   # review without punctuation and numbers
  review=re.sub("[^a-zA-Z]",' ',review)
  # converting into lowercase and splitting to eliminate stopwords
  review=review.lower()
  review=review.split()
  # review without stopwords
  swords=set(stopwords.words("english"))  # conversion into set for fast searching
  review=[i for i in review if i not in swords]
  return(" ".join(review))
```

```python
train_x_tum = []
for r in range(len(df["review"])):
    if (r+1)%1000 == 0:
        print("No of reviews processed =", r+1)
    train_x_tum.append(process(df["review"][r]))
```

```python
X=train_x_tum
y=np.array(df["sentiment"])

# train test split
train_x, test_x, y_train, y_test = train_test_split(X,y, test_size = 0.1,random_state=50)
```

```python
# sklearn içinde bulunan countvectorizer fonksiyonunu kullanarak max 5000 kelimelik bag of words
vectorizer=CountVectorizer(max_features=500)

# train verilerimizi feature vektöre matrisine çevirme
train_x=vectorizer.fit_transform(train_x)
```

```python
train_x=train_x.toarray()
```

```python
train_y=y_train
```

```python
train_x.shape, train_y.shape
```

```python
train_y
```

Random Forest Model

```python
r_model=RandomForestClassifier(n_estimators=100,random_state=50)
r_model.fit(train_x, train_y)
```

```python
# Test verilerini feature vektöre matrisine çevirme
test_xx = vectorizer.transform(test_x)
```

```python
test_xx
```

```python
test_xx = test_xx.toarray()
```

```python
r_test_predict = r_model.predict(test_xx)
```

```python
r_test_predict
```

```python
d=roc_auc_score(y_test,r_test_predict)
```

```python
print("Doğruluk oranı : % ", d * 100)
```

Naive Bayes Model

```python
nb_model = MultinomialNB()
nb_model.fit(train_x, train_y)
```

```python
# Test veri kümesi üzerinde tahmin yapma
n_test_predict = nb_model.predict(test_xx)
```
